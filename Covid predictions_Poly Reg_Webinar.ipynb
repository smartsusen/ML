Let's fabricate some data that shows a roughly linear relationship :

## We need a Polynomial Regression Model

We will now build another model using polynomical regression. 

```python
x = np.array(Age)
y = np.array(Salary)



#p4 = np.poly1d(np.polyfit(x, y, 1))
np.polyfit(x, y, 2)
Y_predicted = np.poly1d(np.polyfit(x, y, 2))

print(Y_predicted)
```

```python
import matplotlib.pyplot as plt

xp = np.linspace(0, 100, 100)
plt.scatter(x, y)
plt.plot(xp, Y_predicted(xp), c='r')
plt.show()
```
